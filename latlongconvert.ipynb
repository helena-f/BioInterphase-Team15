```python
# logic for -180 to 180 to 0 to 360 format
def convert_longitude(lon):

    return lon % 360 if lon < 0 else lon # to lat long in 0 to 360 form

# testing
longitude = -170
converted_lon = convert_longitude(longitude)
print(f"Converted Longitude: {converted_lon}")
```

```python
def convert_latitude(lat):
    return lat + 90

# testing
latitudes = [-90, -45, 0, 45, 90]
converted_latitudes = [convert_latitude(lat) for lat in latitudes]

# printing converted latitudes
for original, converted in zip(latitudes, converted_latitudes):
    print(f"Original Latitude: {original} -> Converted Latitude: {converted}")
```

```python
def process_csv(file_path):
    df = pd.read_csv(file_path)

    # Check if both LATITUDE and LONGITUDE columns are there
    if 'LONGITUDE' in df.columns and 'LATITUDE' in df.columns:
        # Apply the conversions
        df['LONGITUDE_converted'] = df['LONGITUDE'].apply(convert_longitude)
        df['LATITUDE_converted'] = df['LATITUDE'].apply(convert_latitude)

        # Replace original LATITUDE and LONGITUDE columns with the converted ones
        df['LONGITUDE'] = df['LONGITUDE_converted']
        df['LATITUDE'] = df['LATITUDE_converted']

        # Drop the extra converted columns
        df = df.drop(columns=['LONGITUDE_converted', 'LATITUDE_converted'])

        # Test print before saving
        print("DataFrame before saving:")
        print(df[['LONGITUDE', 'LATITUDE']].head())

        # Save the updated file as CSV
        output_file_path = '/content/drive/My Drive/cleaned_habsos_data_v3.csv'
        df.to_csv(output_file_path, index=False)
        print(f"Converted data saved to: {output_file_path}")

    else:
        print("LATITUDE and/or LONGITUDE column(s) not found in the dataset.")

# Define file path
file_path = '/content/drive/My Drive/cleaned_habsos_data.csv'

# Process the CSV
process_csv(file_path)
```
