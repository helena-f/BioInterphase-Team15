```python
# Load ARGOS dataset
df_ARGO = pd.read_csv("/content/drive/MyDrive/Team15-Biointerphase/Dataset/SOCCOM_BGC_Argo_Snapshot_Archive_38d9_96df_e9c7.csv", header=0)
df_ARGO.head()
```

```python

# Ensure the 'date' column is in datetime format
df_NCCA['DATE_COL'] = pd.to_datetime(df_NCCA['DATE_COL'])

# Get the earliest and latest date
earliest_date = df_NCCA['DATE_COL'].min()
latest_date = df_NCCA['DATE_COL'].max()

# Print the range of dates
print(f"Date range: {earliest_date} to {latest_date}")
```

```python
# Ensure the 'date' column is in datetime format
df_ARGO = df_ARGO.drop(index=0)
df_ARGO['time'] = pd.to_datetime(df_ARGO['time'],utc=True)
df_ARGO['time'] = df_ARGO['time'].dt.tz_localize(None)
df_ARGO.head()
```

```python
df_ARGO = df_ARGO.rename(columns={'time': 'date'})
df_NCCA = df_NCCA.rename(columns={'DATE_COL': 'date'})
df_NCCA = df_NCCA.rename(columns={'LAT_DD': 'latitude'})
df_NCCA = df_NCCA.rename(columns={'LON_DD': 'longitude'})
print(df_ARGO.columns)
print(df_NCCA.columns)
```

```python
# Get long/lat comparisons
df_ARGO['latitude'] = pd.to_numeric(df_ARGO['latitude'], errors='coerce')
df_ARGO['longitude'] = pd.to_numeric(df_ARGO['longitude'], errors='coerce')

print("df_NCCA: \n")
print("latitude range: " + str(df_NCCA['latitude'].min()) + " " + str(df_NCCA['latitude'].max()) + "\n")
print("longitude range: " + str(df_NCCA['longitude'].min()) + " " + str(df_NCCA['longitude'].max()) + "\n")

print("df_ARGO: \n")
print("latitude range: " + str(df_ARGO['latitude'].min()) + " " + str(df_ARGO['latitude'].max()) + "\n")
print("longitude range: " + str(df_ARGO['longitude'].min()) + " " + str(df_ARGO['longitude'].max()) + "\n")

# Gulf of Mexico, USA Latitude and longitude coordinates are: 25.681137, -89.890137
```

```python
# Combine based on date
from geopy.distance import geodesic
from itertools import product

# Function to calculate distance between two lat/lon pairs
def calculate_distance(row1, row2):
    return geodesic((row1['latitude_ARGO'], row1['longitude_ARGO']), (row2['latitude_NCCA'], row2['longitude_NCCA'])).kilometers

# Set a distance threshold (e.g., 1 km)
threshold_km = 1.0

# Perform an inner merge on the 'date' column
merged_on_date = pd.merge(df_ARGO, df_NCCA, on='date', suffixes=('_ARGO', '_NCCA'))
merged_on_date.columns
```

```python
merged_on_date['latitude_ARGO'] = pd.to_numeric(merged_on_date['latitude_ARGO'], errors='coerce')
merged_on_date['latitude_NCCA'] = pd.to_numeric(merged_on_date['latitude_NCCA'], errors='coerce')
merged_on_date['longitude_ARGO'] = pd.to_numeric(merged_on_date['longitude_ARGO'], errors='coerce')
merged_on_date['longitude_NCCA'] = pd.to_numeric(merged_on_date['longitude_NCCA'], errors='coerce')

# Example: Print 'latitude_ARGO', 'longitude_ARGO', 'latitude_NCCA', and 'longitude_NCCA' columns
# print(merged_on_date[['latitude_ARGO', 'longitude_ARGO', 'latitude_NCCA', 'longitude_NCCA']])
# Define a tolerance level (e.g., within 0.0001 degrees)
tolerance = 190

# # Check if latitudes and longitudes are approximately equal within the tolerance
lat_match = np.isclose(merged_on_date['latitude_ARGO'], merged_on_date['latitude_NCCA'], atol=tolerance)
lon_match = np.isclose(merged_on_date['longitude_ARGO'], merged_on_date['longitude_NCCA'], atol=tolerance)

# Combine the two conditions
approx_match = lat_match & lon_match

# Output the rows where both latitude and longitude approximately match
matching_rows = merged_on_date[approx_match]
print(matching_rows.columns)
print(matching_rows)
```

```python

# Filter based on location proximity
filtered_rows = []
for idx, row in merged_on_date.iterrows():
    dist = calculate_distance(row[['latitude_ARGO', 'longitude_ARGO']], row[['latitude_NCCA', 'longitude_NCCA']])
    if dist <= threshold_km:
        filtered_rows.append(row)

# Convert the list of filtered rows back to a DataFrame
merged_df = pd.DataFrame(filtered_rows)
```
